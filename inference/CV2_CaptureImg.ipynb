Minimum code for capturing an img

```python
img_path = "img/captured_image.jpg"
if os.path.exists(img_path):
    os.remove(img_path)
    print(f"Previous img '{img_path}' has been removed.")

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cap.set(cv2.CAP_PROP_FPS, 30)

status, photo = cap.read()
cv2.imwrite(img_path, photo)
cap.release()

Image(filename=img_path, height=640)
```

Capture an image and measure the time it takes for it
- Remove image if there is a previous one. 
- Capture a new image and store it in img/captured_image.jpg

```python
img_path = "img/captured_image.jpg"
if os.path.exists(img_path):
    os.remove(img_path)
    print(f"Previous img '{img_path}' has been removed.")

cap = cv2.VideoCapture(0)

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
cap.set(cv2.CAP_PROP_FPS, 30)

start_time = time.time()
status, photo = cap.read()
end_time = time.time()
cv2.imwrite(img_path, photo)
cap.release()


elapsed_time = end_time - start_time
print(f"Time taken to capture and save the image: {elapsed_time:.4f} seconds")
Image(filename=img_path, height=640)
```

Capture img with brightness adjust

```python

img_path = "img/captured_image_bright.jpg"
if os.path.exists(img_path):
    os.remove(img_path)
    print(f"Previous img '{img_path}' has been removed.")

cap = cv2.VideoCapture(0)
status, photo = cap.read()

# Adjust brightness
brightness_value = 50  # Increase this value for more brightness
bright_photo = cv2.convertScaleAbs(photo, alpha=1, beta=brightness_value)

# Save the brightened image
cv2.imwrite(img_path, bright_photo)

# Release the camera
cap.release()

Image(filename=img_path, height=640)
```
